```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/Expanded_data_with_more_features.csv")

print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis =1)
df.head()
```

Change wklyStudyHours column

```python
df['WklyStudyHours']=df['WklyStudyHours'].str.replace("5-10","05-10")
df.head()
```

Gender distribution

```python
sns.countplot(data=df,x='ParentEduc')
```

```python
sns.distplot(df['MathScore'])
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analysed that:
The number of females in the data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart:
we have concluded that the education of the parents have a good impact on their scores

Master's degree parents childrens score is more

some high school parents childrens score is less

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship Between Parent's Martial staus and Student's Score ")
plt.show()
```

From the above chart we concluded that:
There is no/negligible impact on the student's score due to their parents Marital status


```python
gb2=df.groupby("WklyStudyHours").agg({'MathScore':'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb2,annot=True)
plt.title("Relationship between WklyStudyHours and Students Score ")
```

From the above chart: we have concluded that the WklyStudyHours have a good impact on their scores

WklyStudyHours is >10 students score is more

WklyStudyHours is < 5 students  score is less

```python
groupA=df.loc[(df['EthnicGroup']== "group A")].count()
groupB=df.loc[(df['EthnicGroup']== "group B")].count()
groupC=df.loc[(df['EthnicGroup']== "group C")].count()
groupD=df.loc[(df['EthnicGroup']== "group D")].count()
groupE=df.loc[(df['EthnicGroup']== "group E")].count()
l=['group A','group B','group C','group D','group E']
mylist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
print(mylist)
plt.title("Distribution of Ethnic groups")
plt.pie(mylist, labels=l,autopct="%1.2f%%")
plt.legend()
plt.show()
```
